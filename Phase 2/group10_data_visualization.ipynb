### 3. Annotations
* Annotation type: xml
* Essential Annotations:
    * filename
    * object
    * class: "name"
    * bounding box: "bndbox"

Below is the sample of the annotations from one of the file

#### 3.1 Object
* The obect refers to one face in the image with its properties nested. 

#### 3.2 Bounding box:
* The bounding box with the tag name 'bndbox' contains 4 nested tags which refer to the x and y co-ordinates of the object, i.e face.

```python
im = imread('./images/maksssksksss2.png')
fig = figure(0, (12, 12))
fig.add_subplot(1,2,1)
imshow(im); title('Original Image');
im = im[31:65, 333:366, :]
fig.add_subplot(1,2,2)
imshow(im); title('Cropped image from annotations');
```
